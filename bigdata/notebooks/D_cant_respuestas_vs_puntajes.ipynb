# Stack Overflow Posts Score vs Answer Count Anaylsis
Simple notebook for finding a relationship between Score and Answer Count for Stack Overflow Posts.
## Imports
Let's import some needed packages.

```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing data
Now import the .csv file generated by our MapReduce script "D_cant_respuestas_vs_puntajes.py" in "src" folder.

```python
#CSV file path
CSV_PATH = '../output/D_cant_respuestas_vs_puntajes.csv'
#Read csv
df = pd.read_csv(CSV_PATH, index_col=0)
#Check dataframe
df
```

## Visualize data
We can see is a really small dataset showing the average score for each answer count interval.
The best way to visualize this is with a bar plot.

```python
#Order intervals
ordered_intervals = ['0-1', '1-2', '2-5', '5-10', '10-25', '25-50', '50-100', '100-200']
#Plot a barplot
plt.rcParams["figure.figsize"] = (10,5)
ax = sns.barplot(x="AnswerCountInterval", y="AverageScore", data=df, order=ordered_intervals)
ax.set_title("Stack Overflow Posts Score by Answer Count", fontsize=20)
ax.set_ylabel("Average Score")
ax.set_xlabel("Answer Count Interval")
plt.show()
```

It seems like there is a strong relationship between the 2 variables, but the intervals are handling different scales.

Let's use a logarithmic scale for Average Score in order to better visualize this (since Answer Count Interval scale is quite logarithmic already too).

```python
#Plot a barplot with log scale in y axis
plt.rcParams["figure.figsize"] = (10,5)
ax = sns.barplot(x="AnswerCountInterval", y="AverageScore", data=df, order=ordered_intervals)
ax.set_title("Stack Overflow Posts Score by Answer Count", fontsize=20)
ax.set_ylabel("Average Score")
ax.set_xlabel("Answer Count Interval")
ax.set(yscale="log")
plt.show()
```

## Insights
Now it's clear. There is a strong positive correlation between Score and Answer Count for Stack Overflow Posts. It seems like more popular posts (with a lot of answers) are the ones with better scores in this dataset.

We can also see an interesting average score increase for posts with "5-10" answers.
