# Stack Overflow Posts Anaylsis - Top 10 questions with no accepted answer
Simple notebook to analyze the Top 10 Stack Overflow questions that have not an accepted answer yet.
## Imports
Let's import some needed packages.

```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing data
Now import the .csv file generated by our MapReduce script "D_top10_tipos_post_sin_respuestas_aceptadas.py" in "src" folder.

```python
#CSV file path
CSV_PATH = '../output/D_top10_tipos_post_sin_respuestas_aceptadas.csv'
#Read csv
df = pd.read_csv(CSV_PATH, index_col=0)
#Check dataframe
df
```

# Visualize data
It's a really small dataset showing the Answer Count of 10 different posts. The best way to plot this is with a bar plot.

```python
#Ordered IDs
ordered_ids = list(df.Id.values)
#Plot an horizontal bar plot
plt.rcParams["figure.figsize"] = (10,5)
ax = sns.barplot(x="AnswerCount", y="Id", data=df, orient='h', order=ordered_ids)
ax.set_title("Top 10 Posts with no Accepted Answer by Answer Count", fontsize=20)
ax.set_ylabel("Post ID")
#ax.set_yticklabels([])
ax.set_xlabel("Answer Count")
plt.show()
```
