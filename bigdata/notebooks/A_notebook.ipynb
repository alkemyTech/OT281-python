```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Top 10 TAGS with accepted answers

```python
top10 = pd.read_csv('A_top10_tags_accepted_answers.csv')
top10.head()
```

```python
plt.figure(figsize=(10,5))
plt.bar(top10['Tag'], top10['AnswerCount'])
plt.title('Top 10 of Tags that has more Answer Counts')
plt.xlabel("Tags (top 10)')")
plt.xticks(rotation=45)
plt.ylabel('# of total answers
plt.show()
```

### Analysis:
* __*discussion*__ is the number 1 in tags, doubling the second one __*feature-request*__.
* __*support*__ and __*stackoverflow*__ follow as third and fourth, but the half of answers from the second each of them.
* As per the rest, they decrease slowly until ~ 2000 answers.

## Score and word-count per post relationship

```python
word_count = pd.read_csv('A_word_count_score.csv')
word_count.head()
```

```python
word_count['Score'] = word_count['Score'].astype(int)
```

```python
word_count = word_count.sort_values(by=['Score'])
```

```python
plt.figure(figsize=(10,6))
plt.bar(word_count['Score'], word_count['Word count (average)'])
plt.title('Score and word-count per post')
plt.xlabel('Score')
plt.ylabel('Word count')
plt.show()
```

### Analysis
* The majority of scores are below 200.
* Scores upper 200, will be considered outliers.
* There is no correlation between #words vs score.
